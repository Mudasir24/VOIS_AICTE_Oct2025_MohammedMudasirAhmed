```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Airbnb_Open_Data.xlsx')
```

Clean the Data

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(['house_rules', 'license'], axis=1, inplace=True)
```

```python
df.rename(columns = {
'price':'price_$',
'service_fee': 'service fee_$'
}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].astype(int)
```

```python
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
```

```python
df = df.drop(df[df['availability 365'] > 500].index)
```

1. What are the different property types in the Dataset?

```python
property_types = df['room type'].value_counts().to_frame()
property_types
```

```python
room_type_bar = plt.bar(property_types.index, property_types.loc[:,"count"])
plt.bar_label(room_type_bar, labels=property_types.loc[:,"count"],padding=6)
plt.ylim(0, 50000)
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Count of Room Types')
plt.show()
```

2. Which neighborhood group has the highest number of listings?

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
hood_group_bar = plt.bar(hood_group.index, hood_group.loc[:,"count"])
plt.bar_label(hood_group_bar, labels=hood_group.loc[:,"count"],padding=6)
plt.ylim(0, 40000)
plt.xlabel('Neighborhood Group')
plt.ylabel('Count')
plt.title('Count of Neighborhood Groups')
plt.show()
```

3. Which neighborhoods group have the highest average prices for Airbnb listings?


```python
avg_price_hood_group = df.groupby('neighbourhood group')['price_$'].mean().to_frame()
avg_price_hood_bar = plt.bar(avg_price_hood_group.index, avg_price_hood_group.loc[:,"price_$"])
plt.bar_label(avg_price_hood_bar, labels=round(avg_price_hood_group.loc[:,"price_$"],2),padding=10)
plt.ylim(0, 700)
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.title('Average Price by Neighborhood Group')
plt.show()
```

4. Is there a relationship between the construction year of property and price?



```python
df.groupby(df['Construction year'])['price_$'].mean().to_frame().plot()
plt.xlabel('Construction Year')
plt.ylabel('Average Price')
plt.title('Average Price by Construction Year')
plt.show()
```

5. Who are the top 10 hosts by calculated host listing count?
[link text](https://)

```python
hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False).head(10).to_frame()
hosts_bar = plt.bar(hosts.index, hosts.loc[:,"calculated host listings count"])
plt.bar_label(hosts_bar, labels=hosts.loc[:,"calculated host listings count"],padding=10)
plt.ylim(0, 130000)
plt.xlabel('Host Name')
plt.xticks(rotation=90)
plt.ylabel('Calculated Host Listings Count')
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.show()
```

6. Are hosts with verified identities more likely to receive positive reviews?


```python
review = df.groupby('host_identity_verified')['review rate number'].mean().to_frame()
review_bar = plt.bar(review.index, review.loc[:,"review rate number"])
plt.bar_label(review_bar, labels=round(review.loc[:,"review rate number"],2),padding=10)
plt.ylim(0, 5)
plt.xlabel('Host Identity Verified')
plt.ylabel('Review Rate Number')
plt.title('Review Rate Number by Host Identity Verified')
plt.show()
```

7. Is there a correlation between the price of a listing and its service fee?


```python
df['price_$'].corr(df['service fee_$'])
```

```python
sns.regplot(x='price_$', y='service fee_$', data=df)
plt.xlabel('Price')
plt.ylabel('Service Fee')
plt.title('Correlation between Price and Service Fee')
plt.show()
```

8. What is the average review rate number (e.g., stars) for listings, and does it vary based on the neighborhood group and room type?


```python
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame()
ARRN
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood group', y='review rate number', hue='room type', data=df)
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Review Rate Number')
plt.title('Average Review Rate Number by Neighborhood Group and Room Type')
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.show()
```

9. Are hosts with a higher calculated host listings count more likely to maintain higher availability throughout the year?

```python
sns.regplot(x='calculated host listings count', y='availability 365', data=df)
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Availability 365')
plt.title('Correlation between Calculated Host Listings Count and Availability 365')
plt.show()
```
